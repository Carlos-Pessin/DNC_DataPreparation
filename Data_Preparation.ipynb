```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv('DataPrepFinal.csv')
df_c_raw = pd.read_csv('campaign.csv')
df_i_raw = pd.read_csv('invested.csv')
```

```python
df_raw.head()
```

```python
df_raw.info()
```

```python
# removing unnecessary data
df = df_raw.drop(columns=['Unnamed: 0', 'name', 'category', 'goal', 'pledged', 'usd pledged'])
```

```python
df.head()
```

```python
# removing 'USD' string from column usd_goal_real
df['usd_goal_real'] = df['usd_goal_real'].apply(lambda x: x.strip('USD '))

# removing hours/minutes information from column launched
df['launched'] = df['launched'].str.replace(' \d\d:\d\d', '', regex=True)   # regex looks for a pattern to be replaced, where '\d' is a number 0-9
```

```python
df.head()
```

```python
df.info()
```

```python
# correcting formats
df['usd_goal_real'] = df['usd_goal_real'].astype('int64')
df['deadline'] = pd.to_datetime(df['deadline'], format='%d/%m/%y')
df['launched'] = pd.to_datetime(df['launched'], format='%d/%m/%y')
```

```python
df.info()
```

```python
# creating a column that calculates time difference between launched and deadline dates
df['time_range'] = (df['deadline'] - df['launched']).dt.days    # dt.days to make it an integer number
```

```python
df.head()
```

```python
df.info()
```

```python
df_c_raw.head()
```

```python
df_c_raw['Text Description'].unique()
```

```python
# removing column because all values are equal
df_c = df_c_raw.drop(columns=['Text Description'])
```

```python
df_c.head()
```

```python
df_c.info()
```

```python
# joining datasets
df = df.merge(df_c, how='right', on=['ID'])
```

```python
df.head()
```

```python
df_i_raw.head()
```

```python
# one hot encoding gor backedLocation so we can aggregate lines with equal ID
df_i = pd.get_dummies(df_i_raw, columns=['backedLocation'])
df_i.head()
```

```python
df_i.info()
```

```python
# grouping lines
df_i = df_i.groupby(by=['ID']).agg({'age':'mean', 'backedLocation_BR':'sum', 'backedLocation_GBK':'sum', 'backedLocation_US':'sum'})
```

```python
df_i.head()
```

```python
df = df.merge(df_i, how='right', on=['ID'])
```

```python
df = df.rename(columns={'age':'investors_mean_age', 'backedLocation_BR':'BR_investors_count', 'backedLocation_GBK':'GBK_investors_count', 'backedLocation_US':'US_investors_count'})
```
